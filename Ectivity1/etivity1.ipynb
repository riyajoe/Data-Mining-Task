Most of the clients are contacted during the month of February have not been subscribed whereas customers who were contacted during July ,almost all were subscribed .
From client’s previous campaigns it can be identified that clients who were contacted less than 5 times had been subscribed out of which there were small portion of  failures were observed and huge portion of the subscribed had non existent campaigns 
Most of the subscribed clients campaigns were conducted through cellular and most of them were contacted around 25 to 20 times before a campaign. Most of the subscribed customers where contacted within a span of 200 days .The duration for campaigns performed less than 6 times were more than that of campaigns conducted greater than 6 times having cellular as mode of contact.
